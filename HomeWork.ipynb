###### Import relevant libraries

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
import matplotlib.pyplot as plt
```

###### URL of Pokemon DataSet

```python
html = 'https://gist.githubusercontent.com/simsketch/1a029a8d7fca1e4c142cbfd043a68f19/raw/bd584ee6c307cc9fab5ba38916e98a85de9c2ba7/pokemon.csv'
```

###### My Extraction Logic

```python
page = urlopen(html)
data = BeautifulSoup(page, 'html.parser')
name_box = data.findAll('div', attrs={'class': 'col-md-3 col-sm-3'})
```

```python
for i in range(len(name_box)):
    data = name_box[i].text.strip()
```

```python
data
```

###### Save extracted data into desired folder

```python
with open("C:/Users/username/Downloads/Darbs Texxt.csv", "w") as carCountsCsv:
    for word in data :
        resultCar = str(word)
        print(resultCar)
        carCountsCsv.write(resultCar)
```
