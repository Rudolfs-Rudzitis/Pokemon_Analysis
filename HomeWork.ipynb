### 1. Collect information about Pokemons

```python
page = urlopen(html)
data = BeautifulSoup(page, 'html.parser')
name_box = data.findAll('div', attrs={'class': 'col-md-3 col-sm-3'})
```

```python
for i in range(len(name_box)):
    data = name_box[i].text.strip()
```

## Exploratory Data Analysis (EDA)

###### Summary of the variables found in the datasets

```python
# # ID for each pokemon
# Name - Name of pokemon
# Type 1 - Each pokemon has a type, this determines weakness/resistance to attacks
# Type 2 - Some pokemon are dual type and have 2
# Total - sum of all stats that come after this, a general guide to how strong a pokemon is
# HEIGHT - height of the pokemon expressed in 100g (so 69 would mean 6.9kg)
# WEIGHT - weight of the pokemon expressed in 10cm (so 0.7 would mean 70cm)
# HP - hit points, or health, defines how much damage a pokemon can withstand before fainting
# ATK - the base modifier for normal attacks (eg. Scratch, Punch)
# DEF - the base damage resistance against normal attacks
# SP_ATK - special attack, the base modifier for special attacks (e.g. fire blast, bubble beam)
# SP_DEF - the base damage resistance against special attacks
# SPD - determines which pokemon attacks first each round
# TOTAL - sum of all stats
```
